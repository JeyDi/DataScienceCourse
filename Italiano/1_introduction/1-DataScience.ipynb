## What is Data Science?

Data Science, nota anche come: scienza basata sui dati, è un campo interdisciplinare che riguarda metodi, processi e sistemi scientifici per estrarre conoscenze o intuizioni dai dati in varie forme, sia strutturate che non strutturate.  

Ci sono molte definizioni sulla scienza dei dati, per esempio:  
- La scienza dei dati è un "concetto per unificare la statistica, l'analisi dei dati e i loro metodi correlati" al fine di "comprendere e analizzare i fenomeni reali" con i dati. - Chikio Hayashi  

Ed è impossibile non menzionare anche il: Machine Learning
- L'apprendimento automatico è un "campo di studio che dà ai computer la possibilità di imparare senza essere esplicitamente programmati" - Arthur Samuel

Quindi l'obiettivo dell'apprendimento automatico (Machine Learning) è:
- programmare i computer per usare dati di esempio o esperienze passate per risolvere un dato problema

Quindi usiamo Data Science e Machine Learning per consentire ai computer di aiutarci a capire o a fare previsioni sul mondo.

Quando parliamo di scienza dei dati, parliamo anche di **Analisi dei dati**  
**L'analisi dei dati** --> è un processo attraverso il quale si indaga su grandi quantità di dati per determinare le conlusioni e per supportare il processo decisionale, quindi dobbiamo guardare a ciò che i dati stanno cercando di dirci.

Ma in pratica? What do you do in a Data Science project or work? 

- Capire come lavorare con i dati (estrazione, pulizia e organizzazione) da molte fonti diverse con diversi tipi di dati (dati strutturati, non strutturati)

## Che cos'è l'Intelligenza Artificiale?

Non esiste una definizione giusta, corretta o completa, spesso si evolvono e cambiano nel tempo...
È molto difficile definire cosa sia Intelligenza Artificiale perchè allo stesso tempo è difficile definire cosa sia Intelligenza.

Ci sono inoltre tanti retaggi della fantascienza, che però faticano ad avere un riscontro con il mondo reale.

Proviamo comunque a trovare delle definizioni per definire meglio il contesto, utilizzando alcune definizioni autorevoli che si possono trovare nel web:

Merriam-Webster definisce l'intelligenza artificiale in questo modo:

    Un ramo dell'informatica che si occupa della simulazione del comportamento intelligente nei computer.
    La capacità di una macchina di imitare il comportamento umano intelligente.
      
L'Enciclopedia Britannica afferma, 

    "Intelligenza artificiale (AI), la capacità di un computer digitale o di un robot controllato da un computer di svolgere compiti comunemente associati agli esseri intelligenti". 
    Gli esseri intelligenti sono quelli in grado di adattarsi al cambiamento delle circostanze.


L'Unione europea, con una commissione specifica, ha creato le linee guida etiche europee per un'IA affidabile.
https://ec.europa.eu/digital-single-market/en/news/ethics-guidelines-trustworthy-ai  
https://ec.europa.eu/futurium/en/ai-alliance-consultation  

la definizione che hanno dato è la seguente:

    "Intelligenza artificiale (IA) si riferisce a sistemi che mostrano un comportamento intelligente analizzando il loro ambiente e intraprendendo azioni - con un certo grado di autonomia - per raggiungere obiettivi specifici.
    I sistemi basati sull'intelligenza artificiale possono essere puramente software, agendo nel mondo virtuale (ad esempio assistenti vocali, software di analisi delle immagini, motori di ricerca, sistemi di riconoscimento vocale e facciale) o l'intelligenza artificiale può essere incorporata in dispositivi hardware (ad esempio robot avanzati, auto autonome, droni o applicazioni Internet of Things)".


Il mantra dell'IA è: Ciò che sembra facile è in realtà difficile... e ciò che sembra difficile è in realtà facile

Per dare una definizione corretta, è forse più utile ragionare per applicazioni concrete e per modi di risolvere determinati problemi.
Alcuni esempi di applicazione di algoritmi di apprendimento automatico sono:
- Sistemi di macchine a guida autonoma
- Sistemi di analisi di anomalie in tempo reale (es: frodi bancarie, sensori, ...)
- Agenti intelligenti e simulazione
- Algoritmi in grado di giocare a giochi da tavolo o elettronici meglio degli umani
- Sistemi di diagnosi medica avanzati e automatici
- Analisi del testo automatizzata, sentiment analysis o traduttori automatici
- Chatbot testuali e vocali
- ...

Ci sono però due argomenti principali, due termini che sono estremamente importanti quando si parla di IA:

**Autonomia**
- Si tratta della capacità di eseguire compiti in ambienti complessi senza una guida costante da parte dell'utente.

**Attività**
- Questa è la capacità di migliorare le prestazioni imparando dall'esperienza.

C'è un ciclo di vita per realizzare un progetto, un'applicazione nel campo dell'Intelligenza Artificiale
